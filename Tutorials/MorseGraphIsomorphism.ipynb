## Check if Morse graphs are isomorphic

```python
net_spec = """x : x + y : B E
              y : (~x) y : B E"""

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Number of parameter nodes:', parameter_graph.size())
```

```python
par_index = 6

parameter = parameter_graph.parameter(par_index)

morse_graph1, stg1, graded_complex1 = DSGRN.Blowup.ConleyMorseGraph(parameter)
```

```python
DSGRN.Blowup.PlotMorseGraph(morse_graph1)
```

```python
par_index = 14

parameter = parameter_graph.parameter(par_index)

morse_graph2, stg2, graded_complex2 = DSGRN.Blowup.ConleyMorseGraph(parameter)
```

```python
DSGRN.Blowup.PlotMorseGraph(morse_graph2)
```

Check if the Morse graphs are isomorphic.

```python
%%time
DSGRN.isomorphic_morse_graphs(morse_graph1, morse_graph2)
```
