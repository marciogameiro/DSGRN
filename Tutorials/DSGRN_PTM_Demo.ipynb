# Basic usage

Decay terms are enclosed by `<>` and must be at the beginning of the line in the network specification. The remaining terms must be added to the decay terms. Pairs of PTM edges must be enclosed by `[]`. A decreasing step function is used for varibales preceded by a `~`.

```python
net_spec = 'x : <~z> + y\n y : <x> + ~z\n z : [x, ~y]'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

Once the network is constructed, everything else is exactly the same as in the regular DSGRN.

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('There are', parameter_graph.size(), 'nodes in the parameter graph.')
```

```python
net_spec = 'x : z\n y : <x> + ~z\n z : [x, y]'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('There are', parameter_graph.size(), 'nodes in the parameter graph.')
```

When drawing the network, one can select how the PTM pairs are joined. The options are `p` (for point - the default), `c` (for circel), and `s` (for sqaure). This only affects the drawing the of the network. The DSGRN computations remain the same.

```python
net_spec = 'x : z\n y : <x> + ~z\n z : [x, y]'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network, 'p')
```

```python
net_spec = 'x : z\n y : <x> + ~z\n z : [x, y]'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network, 'c')
```

```python
net_spec = 'x : z\n y : <x> + ~z\n z : [x, y]'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network, 's')
```

# PTM network notation

One can use the command `DrawNetworkPTM` to draw the network using the PTM notation. In this notation inactive edges (the default) are shown with a filled circle and active edges are shwon with a empty circle. An active edge is indicated by an additional pair of `[]` enclosing the variable. Only the first element of a PTM pair is allowed to be active/inactive.

```python
ptm_net_spec = 'x : z\n y : <x> + ~z\n z : [x, y]'

DSGRN.DrawNetworkPTM(ptm_net_spec)
```

```python
ptm_net_spec = 'x : z\n y : <x> + ~z\n z : [[x], y]'

DSGRN.DrawNetworkPTM(ptm_net_spec)
```

If the network specification is in the PTM format, the command `NetworkPTM` can be used to construct the network.

```python
network = DSGRN.NetworkPTM(ptm_net_spec)
```

Also in this case, once the network is constructed everything else is exactly the same as in the regular DSGRN.

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('There are', parameter_graph.size(), 'nodes in the parameter graph.')
```

One can get the regular DSGRN network specification from the PTM netwoek specification with the command `SpecPTM2Spec`. In fact `NetworkPTM(ptm_net_spec)` is just a call to `Network(SpecPTM2Spec(ptm_net_spec))`.

```python
DSGRN.SpecPTM2Spec(ptm_net_spec)
```
