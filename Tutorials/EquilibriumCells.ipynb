# DSGRN Equilibrium Cells

```python
import DSGRN
```

```python
net_spec = 'X : ~Y X \n Y : ~X Y'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Parameter graph size:', parameter_graph.size())
```

```python
par_index = 154

parameter = parameter_graph.parameter(par_index)
```

The command `EquilibriumCells` returns the equilibirum cells for a given parameter. Each equilibrium cell is given by a pair, where the first entry of the pair represents the vertex with minimum coordinates of the cell and the second entry gives the shape of the cell by indicating which coordinates have extent (an entry 1 indicates that the cell have extent at that coordinate). For example, `[(3, 0), (1, 1)]` represents that 2-dimensional cell [3, 4] x [0, 1], `[(3, 1), (0, 0)]` represents the vertex (3, 1), and `[(2, 3), (0, 1)]` represents the 1-dimensional cell {2} x [3, 4].

```python
DSGRN.EquilibriumCells(parameter)
```

One can specify the the types of equilibrium cells to compute with the optional argument `eqtype`. The default value is `all` and it returns all equilibrium cells. Other options are `topdim` to return only top-dimensional cells and `nontop` to return only non top-dimensional cells.

```python
DSGRN.EquilibriumCells(parameter, eqtype='all')
```

With the option `topdim` only the minimum coordinates of the cells are returned.

```python
DSGRN.EquilibriumCells(parameter, eqtype='topdim')
```

```python
DSGRN.EquilibriumCells(parameter, eqtype='nontop')
```

One can also specify the output format with the optional argument `eqformat`. The options are `coords` (the default) which returns the minimum coordinates coordinates and shape and `index` which returns the indices of the cells in the phase space cubical complex. 

```python
DSGRN.EquilibriumCells(parameter, eqtype='all', eqformat='coords')
```

```python
DSGRN.EquilibriumCells(parameter, eqformat='index')
```
