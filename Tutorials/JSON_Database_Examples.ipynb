# DSGRN JSON Database

```python
import DSGRN
```

```python
net_spec = 'x : ~y x : E \n y : ~x y : E'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Parameter graph size:', parameter_graph.size())
```

The command `SaveDatabaseJSON` computes the Morse graphs and Morse sets and saves the result to a JSON database for visualization (only available for 2D and 3D networks for now). One can specify the parameter nodes to save with the optional argument `param_indices` (if not specified the it saves all parameter nodes).

```python
# File name to save database
database_fname = 'database_2D_example_1.json'

# Parameter indices to save
param_indices = [0, 1, 3, 7, 9, 2, 157]

DSGRN.SaveDatabaseJSON(network, database_fname, param_indices)
```

```python
# File name to save database
database_fname = 'database_2D_example_2.json'

# Save all parameter nodes
DSGRN.SaveDatabaseJSON(network, database_fname)
```

We can select the color of (some) parameter indices by passing a dictionary of colors (in hex format).

```python
# File name to save database
database_fname = 'database_2D_example_3.json'

# Parameter indices to save
param_indices = [0, 1, 3, 7, 9, 2, 157]

# Define a dictionary with vertex colors
# The remaining vertices get the default color
verts_colors = {}

verts_colors[0] = "#1E8449"
verts_colors[3] = "#1E8439"
verts_colors[9] = "#1E8429"

DSGRN.SaveDatabaseJSON(network, database_fname, param_indices=param_indices, verts_colors=verts_colors)
```

Two other optional arguments are `eq_cells` and `thres_type`. If `eq_cells` is set to True then equilibrium cells are included in the databse (the default is False). The argument `thres_type` indicates the threshold format to output parameter partial orders. The options are `T` to use T to represent the thresholds and (the default) `''` to use `tj` to represent the thresholds.

```python
# File name to save database
database_fname = 'database_2D_example_4.json'

# Parameter indices to save
param_indices = [0, 1, 3, 7, 9, 2, 157]

DSGRN.SaveDatabaseJSON(network, database_fname, param_indices=param_indices, eq_cells=True, thres_type='')
```

```python
# File name to save database
database_fname = 'database_2D_example_5.json'

# Parameter indices to save
param_indices = [0, 1, 3, 7, 9, 2, 157]

DSGRN.SaveDatabaseJSON(network, database_fname, param_indices=param_indices, thres_type='T')
```

## A 3D Example

```python
net_spec = 'X0 : X1 + X2 : E\n X1 : X0 : E\n X2 : X1 + X2 : E'

network = DSGRN.Network(net_spec)

DSGRN.DrawGraph(network)
```

```python
parameter_graph = DSGRN.ParameterGraph(network)

print('Parameter graph size:', parameter_graph.size())
```

```python
# File name to save database
database_fname = 'database_3D_example.json'

DSGRN.SaveDatabaseJSON(network, database_fname, eq_cells=True)
```
