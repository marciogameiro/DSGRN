```python
import DSGRN
```

```python
network = DSGRN.Network('X1 : (X1+X2)(~X3)\n X2 : (X1)\n X3 : (X1)(~X2)')
```

```python
DSGRN.DrawGraph(network)
```

```python
parametergraph = DSGRN.ParameterGraph(network)
```

```python
print("There are " + str(parametergraph.size()) + " nodes in the parameter graph.")
```

```python
par_index = 20
```

```python
# Compute original DSGRN adjacencies
parametergraph.adjacencies(par_index)
```

```python
# Compute original DSGRN adjacencies
parametergraph.adjacencies(par_index, 'pre')
```

```python
# Compute DSGRN adjacencies for a fixed order parameter
parametergraph.adjacencies(par_index, 'fixedorder')
```

```python
# Compute DSGRN adjacencies for co-dim 1 order parameters
parametergraph.adjacencies(par_index, 'codim1')
```

```python
# Print the partial orders defining the parameter
parameter = parametergraph.parameter(par_index)
print(parameter.partialorders('T'))
```

```python
# A shorter version of the partial orders defining the parameter
parameter = parametergraph.parameter(par_index)
print(parameter.partialorders())
```

```python
# Print partial orders for pre order adjacencies
for p_index in parametergraph.adjacencies(par_index):
    parameter = parametergraph.parameter(p_index)
    print(parameter.partialorders())
    print('')
```

```python
# Print partial orders for co-dim 1 order adjacencies
for p_index in parametergraph.adjacencies(par_index, 'codim1'):
    parameter = parametergraph.parameter(p_index)
    print(parameter.partialorders())
    print('')
```

```python
# Print partial orders for fixed orders adjacencies
for p_index in parametergraph.adjacencies(par_index, 'fixedorder'):
    parameter = parametergraph.parameter(p_index)
    print(parameter.partialorders())
    print('')
```

```python
# Print parameters for pre order adjacencies
parameter = parametergraph.parameter(par_index)
print(parameter)
print('')

for p_index in parametergraph.adjacencies(par_index):
    parameter = parametergraph.parameter(p_index)
    print(parameter)
```

```python
# Print parameters for co-dim 1 order adjacencies
parameter = parametergraph.parameter(par_index)
print(parameter)
print('')

for p_index in parametergraph.adjacencies(par_index, 'codim1'):
    parameter = parametergraph.parameter(p_index)
    print(parameter)
```

```python
# Print parameters for fixed orders adjacencies
parameter = parametergraph.parameter(par_index)
print(parameter)
print('')

# Notice that the orders are all the same of par_index
for p_index in parametergraph.adjacencies(par_index, 'fixedorder'):
    parameter = parametergraph.parameter(p_index)
    print(parameter)
```

Let's see another example

```python
par_index = 790
```

```python
# Compute original DSGRN adjacencies
parametergraph.adjacencies(par_index)
```

```python
# Compute DSGRN adjacencies for a fixed order parameter
parametergraph.adjacencies(par_index, 'fixedorder')
```

```python
# Compute DSGRN adjacencies for co-dim 1 order parameters
parametergraph.adjacencies(par_index, 'codim1')
```

Notice that in this case the fixed order adjacent parameters are the same as the codim1. We can see why by looking at the partial orders defining the parameter. As we can see below there is no threshold swap that produces a co-dim 1 neighbor. 

```python
parameter = parametergraph.parameter(par_index)
print(parameter.partialorders('T'))
```
